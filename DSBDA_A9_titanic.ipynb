```python
#Sex vs Survived
sns.catplot(x="Sex", hue="Survived", kind="count", data=titanic)
plt.show()
```

```python
sns.stripplot(x="Sex", y="Age", data=titanic, jitter=False)
```

```python
#Stripplot - Sex vs Age + Survived
sns.stripplot(x="Sex", y="Age", hue="Survived", data=titanic, jitter=True) #split - False By Default
```

```python
sns.stripplot(x="Sex", y="Age", hue="Survived", data=titanic, jitter=True, split=True) #split - False By Default
```

```python
#Group dataset by Pclass and Survived and then unstack them
group = titanic.groupby(['Pclass', 'Survived'])
pclass_survived = group.size().unstack()

#Heatmap - Color encoded 2D representation of data
sns.heatmap(pclass_survived, annot=True, fmt="d")
plt.savefig("heatmap.jpg")
```

```python
tc = titanic.corr()
sns.heatmap(tc, cmap="coolwarm")
plt.title("titanic.corr()")
```

```python
#Violinplot Displays Distribution of Data across all levels of category
sns.violinplot(y="Age", x="Sex", hue="Survived", data=titanic, split=True)
```

```python
#FactorPlot
#Adding FamilySize
titanic["FamilySize"] = 0
titanic["FamilySize"] = titanic["Parch"] + titanic["SibSp"]

#Adding Alone
titanic["Alone"] = 0
titanic.loc[titanic.FamilySize==0, "Alone"] = 1
 
# Factorplot for Family_Size
sns.factorplot(x ='FamilySize', y ='Survived', data = titanic)
 
# Factorplot for Alone
sns.factorplot(x ='Alone', y ='Survived', data = titanic)
```
